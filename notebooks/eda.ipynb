**Ahora vamos con las features que se prueban aqui y luego van a limpio**

Primero creamos una columna nueva en el DataFrame

```python
# Feature 1: total de noches
df_clean["total_nights"] = (
    df_clean["stays_in_weekend_nights"] + 
    df_clean["stays_in_week_nights"]
)

# Feature 2: total de huéspedes
df_clean["total_guests"] = (
    df_clean["adults"] + 
    df_clean["children"].fillna(0) + 
    df_clean["babies"]
)

# Feature 3: categoría de antelación
def categorize_lead_time(x):
    if x <= 30:
        return "short"
    elif x <= 90:
        return "medium"
    else:
        return "long"

df_clean["lead_time_category"] = df_clean["lead_time"].apply(categorize_lead_time)
```

```python
df_clean["lead_time_category"].value_counts()
```

Siento long reserva realizada con más de 90 días, short reserva con menos de 30 días, y medium reserva de entre 30 y 90 días
